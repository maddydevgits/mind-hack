```python
import pandas as pd

dataset=pd.read_csv(
    '../datasets/Hyderabad.csv'
)
```

```python
locations=[]
for i in dataset.Location:
    if i not in locations:
        locations.append(i)
```

```python
locations
```

```python
print(len(locations))
```

```python
n=[]
for i in range(1,len(locations)+1):
    n.append(i)
```

```python
print(n)
```

```python
dataset['Location'].replace(locations,
                        n, inplace=True)
```

```python
dataset.head()
```

```python
dataset
```

```python
import numpy as np
def wrangle(df):
    df.replace(9, np.nan, inplace=True)
    df.dropna(inplace=True)
    df.drop(columns = ['Wifi', 'Wardrobe'], inplace=True)
    maskprice = df["Price"]< 100000000
    df = df[maskprice]
    return df

dataset=wrangle(dataset)
```

```python
X=dataset.iloc[:,1:]
```

```python
X=X.values
```

```python
Y=dataset.iloc[:,0]
```

```python
Y=Y.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train,Y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(X_train,Y_train)
```

```python
from sklearn.metrics import r2_score
Y_pred=regressor.predict(X_test)
print(r2_score(Y_pred,Y_test))
```

```python
Y_pred1=rf.predict(X_test)
print(r2_score(Y_pred1,Y_test))
```

```python
Y_pred2=dt.predict(X_test)
print(r2_score(Y_pred2,Y_test))
```
